```python
import numpy as np
import pandas as pd
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
```

```python
y = lambda x: np.exp(2*x) + np.sin(5*x-np.pi/4)
dy_dx = lambda x: 2*np.exp(2*x) + 5 * np.cos(5*x-np.pi/4)
dy2_dx2 = lambda x: 2*2*np.exp(2*x) - 5 * 5 * np.sin(5*x-np.pi/4)
dy3_dx3 = lambda x: 2*2*2*np.exp(2*x) - 5 * 5 * 5 * np.cos(5*x-np.pi/4)

a = -3
b = 3
```

##   

```python
def lagranz(x,y,t):
    z=0
    for j in range(len(y)):
        p1=1
        p2=1
        for i in range(len(x)):
            if i==j:
                p1=p1*1
                p2=p2*1   
            else: 
                p1=p1*(t-x[i])
                p2=p2*(x[j]-x[i])
    z=z+y[j]*p1/p2
    return z
```

```python
plt.figure(figsize = (8, 8))
x_plot=np.linspace(-3, 3, 100)


n=26
x_th = np.linspace(-3, 3, n)
y_th = y(x_th)
plt.plot(x_th, y_th,'o', color = 'brown')
ynew = [lagranz(x_th, y_th, i) for i in x_plot]
plt.plot(x_plot,ynew, color = 'brown')

display(pd.DataFrame({'  ': zip(x_th, y_th), 
                      '': np.around(y_th - [lagranz(x_th, y_th, i) for i in x_th], 5)}))

n=10
x_th = np.linspace(-3, 3, n)
y_th = y(x_th)
plt.plot(x_th, y_th,'o', color = 'green')
ynew = [lagranz(x_th, y_th, i) for i in x_plot]
plt.plot(x_plot,ynew, color = 'green')

display(pd.DataFrame({'  ': zip(x_th, y_th), 
                      '': np.around(y_th - [lagranz(x_th, y_th, i) for i in x_th], 5)}))

n=5
x_th = np.linspace(-3, 3, n)
y_th = y(x_th)
plt.plot(x_th, y_th,'o', color = 'red')
ynew = [lagranz(x_th, y_th, i) for i in x_plot]
plt.plot(x_plot,ynew, color = 'red')

display(pd.DataFrame({'  ': zip(x_th, y_th),
                      '': np.around(y_th - [lagranz(x_th, y_th, i) for i in x_th], 5)}))

n=4
x_th = np.linspace(-3, 3, n)
y_th = y(x_th)
plt.plot(x_th, y_th,'o', color = 'orange')
ynew = [lagranz(x_th, y_th, i) for i in x_plot]
plt.plot(x_plot,ynew, color = 'orange')

display(pd.DataFrame({'  ': zip(x_th, y_th),
                      '': np.around(y_th - [lagranz(x_th, y_th, i) for i in x_th], 5)}))

n=3
x_th = np.linspace(-3, 3, n)
y_th = y(x_th)
plt.plot(x_th, y_th,'o', color = 'blue')
ynew = [lagranz(x_th, y_th, i) for i in x_plot]
plt.plot(x_plot,ynew, color = 'blue')

display(pd.DataFrame({'  ': zip(x_th, y_th),
              '': np.around(y_th - [lagranz(x_th, y_th, i) for i in x_th], 5)}))

plt.grid(True)
plt.show()
```

```python
def get_result_table(n):   
    ind = np.arange(1, 6)
    
    if n == 5:
        h = (b-a)/n
        x_th  = a + (ind - 0.5)*h

    elif n == 25:
        ind = 5*(ind - 1) + 3
        h = (b-a)/n
        x_th  = a + (ind - 0.5)*h
                
    elif n == 125:
        ind = 25*(ind - 1) + 13
        h = (b-a)/n
        x_th  = a + (ind - 0.5)*h
    
    print('h = ', h, ' n = ', n)

    y_th = y(x_th)    

    
    h_s = h/2
    mas_i = np.array([[1, h_s, h_s**2, h_s**3],
                      [0, 1, 2*h_s, 3*h_s**2],
                      [0, 0, 2, 6*h_s],
                      [0, 0, 0, 6]])


    spline = np.array([])
    for x in x_th:
        B_i = np.array([y(x_th), dy_dx(x_th), dy2_dx2(x_th), dy3_dx3(x_th)])

        #  
        coefs = np.linalg.solve(mas_i, B_i)
        # ------------

        spline = np.hstack([spline, coefs[0] + coefs[1]*h_s + coefs[2]*h_s**2 + coefs[3]*h_s**3])


    return spline, y_th, np.abs(spline - y_th)
```

```python
n = 10
h = (b - a)/n
ind = np.arange(1, 11)
x_th = a + ind*h
y_th = y(x_th)

a0, a1 = np.linalg.solve(np.array([[11, np.sum(x_th)],
                   [np.sum(x_th), np.sum(x_th**2)]]), np.array([np.sum(y_th), np.sum(x_th*y_th)]))

phi1 = a0 + a1*x_th

a0, a1, a2 =  np.linalg.solve(np.array([[11, np.sum(x_th), np.sum(x_th**2)],
                                        [np.sum(x_th), np.sum(x_th**2), np.sum(x_th**3)],
                                        [np.sum(x_th**2), np.sum(x_th**3), np.sum(x_th**4)]]), 
                              np.array([np.sum(y_th), np.sum(x_th*y_th), np.sum(y_th*x_th**2)]))

phi2 = a0 + a1*x_th + a2*x_th**2

a0, a1, a2, a3 =  np.linalg.solve(np.array([[11, np.sum(x_th), np.sum(x_th**2), np.sum(x_th**3)],
                                        [np.sum(x_th), np.sum(x_th**2), np.sum(x_th**3), np.sum(x_th**4)],
                                        [np.sum(x_th**2), np.sum(x_th**3), np.sum(x_th**4), np.sum(x_th**5)],
                                        [np.sum(x_th**3), np.sum(x_th**4), np.sum(x_th**5), np.sum(x_th**6)]]), 
                              np.array([np.sum(y_th), np.sum(x_th*y_th), np.sum(y_th*x_th**2), np.sum(y_th*x_th**3)]))

phi3 = a0 + a1*x_th + a2*x_th**2 + a3*x_th**3
```
