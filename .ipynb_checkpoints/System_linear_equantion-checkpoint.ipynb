```python
# band = True, если матрица трехдиагональная, если нет, то band = False
# norm = True, если матрица нормированна, если нет, то band = False
# size_matrix - размер матрицы
def create_matrix(size_matrix = 10, band = False, norm = False):
    A = np.eye(size_matrix)*10 + np.random.randint(1, 10, size = (size_matrix, size_matrix))   
    if band: 
        A = np.zeros((size_matrix, size_matrix))
        A[np.arange(size_matrix-1), np.arange(1, size_matrix)] += np.random.randint(1, 20)
        A[np.arange(size_matrix), np.arange(size_matrix)] += np.random.randint(1, 20)
        A[np.arange(1, size_matrix), np.arange(size_matrix-1)] += np.random.randint(1, 20)
    if norm:
        A += np.eye(size_matrix)*90
    X = np.random.randint(1, 10, size = (1, size_matrix))
    B = A.dot(X.T)
    return A, X, B
```

## Метод Гаусса



```python
# matr_coef - матрица А размерностью n x n
# separ_column - столбец B размерностью n x 1
def method_Gauss(matr_coef, separ_column): 
    n = separ_column.shape[0]
    matr_coef = np.hstack([matr_coef, separ_column])  
    # прямой ход 
    for k in range(n):
        div = matr_coef[k, k]
        for i_str in range(k, n+1):
            matr_coef[k, i_str] /= div
        
        for j_str in range(k+1, n):
            matr_coef[j_str, :] -= matr_coef[j_str, k] * matr_coef[k, :]

    # обратный ход
    i = n - 1
    while i > 0:
        j = i - 1
        while j >= 0:
            matr_coef[j][n] = matr_coef[j][n] - matr_coef[j][i] * matr_coef[i][n] 
            j -= 1
        i -= 1    
    return matr_coef[:, n]
```

```python
start = time.time()
A, X, B = create_matrix(1000)
display(A, B)
x_d = method_Gauss(A, B)
end = time.time()
display(abs(X - x_d).max(), abs(X - x_d).min())
display(abs(B.T - A.dot(x_d.T)).max(), abs(B.T - A.dot(x_d.T)).min())
display(end - start)
```

# Метод прогонки

```python
def sweep_method(size):
    start = time.time()
    A, X, B = create_matrix(size, 1)
    X_diff = TDMA(A, B)
    end = time.time()
    dB = abs(B - A.dot(X_diff.reshape(size, 1)))
    dX = abs(X - X_diff)
    return A, X, B, dX.min(), dX.max(), dB.min(), dB.max(), end - start
```

```python
# n - число уравнений (строк матрицы)
# b - диагональ, лежащая над главной (нумеруется: [0;n-2])
# c - главная диагональ матрицы A (нумеруется: [0;n-1])
# a - диагональ, лежащая под главной (нумеруется: [1;n-1])
# B - правая часть (столбец)
# x - решение, массив x будет содержать ответ
def TDMA(A,B):
    a = np.hstack([0, A.diagonal(-1)])
    b = np.hstack([A.diagonal(1), 0])
    c = A.diagonal()
    
    n = B.shape[0]
    P_mas = np.zeros(n)
    Q_mas = np.zeros(n)    
    P_mas[0] = -b[0] / c[0]
    Q_mas[0] = B[0] / c[0]
    x = np.zeros(n)

    for i in range(1, n):
        P_mas[i] = -b[i]/(a[i]*P_mas[i-1] + c[i])
        Q_mas[i] = (B[i] - a[i]*Q_mas[i-1])/(a[i]*P_mas[i-1] + c[i])
    
    x[n-1] = Q_mas[n - 1]
    for i in range(n-1, 0, -1):
        x[i - 1] = P_mas[i - 1]*x[i] + Q_mas[i - 1]

    return x
```

```python
sweep_method(1000)
```

# Метод простых итераций

```python
def iterations(eps, A, B, X):
    n = A.shape[0]
    A_new = np.hstack([A, B])
    for i in range(n):
        coef = A_new[i, i]
        A_new[i, :] /= coef
        
    A = A_new[:, :n]
    B = A_new[:, n].reshape(-1, 1)
    A_new = np.eye(n) - A
    X_0 = np.vstack([B.T, B.T + A_new.dot(B).T])
    i = 1
    while 1:
        X_0 = np.vstack([X_0, B.T + A_new.dot(X_0[i, :]).T])
        i += 1
        if abs(X_0[i, :] - X).sum()  < n*eps:
            break
    return X_0[i, :], i

def iter_method(eps, n):
    start = time.time()
    A, X, B = create_matrix(n, 0, 1)
    X_diff, n_iter = iterations(eps, A, B, X)
    dB = abs(B - A.dot(X_diff.reshape(n, 1)))
    dX = abs(X - X_diff)
    end = time.time()
    return dX.min(), dX.max(), dB.min(), dB.max(), n_iter, end-start    
```

```python
for i in [1e-4, 1e-8, 1e-12]:
    for j in [10, 20]:
        dX_min, dX_max, dB_min, dB_max, n_iter, time_ = iter_method(i, j)
        print(i, j)
        print('dX_min, dX_max, dB_min, dB_max, n_iter, time', dX_min, dX_max, dB_min, dB_max, n_iter, time_)
```
