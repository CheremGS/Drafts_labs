```python
import numpy as np
import timeit
import matplotlib.pyplot as plt
import sympy as sp
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
pol_3 = lambda x: (x+4)*(x+2)*(x-4)
df_dx_pol3 = lambda x: 3*x**2 + 4*x - 16

transcend = lambda x: np.sin(2*x) + 5**x
df_dx_transcend = lambda x: 2*np.cos(2*x) + np.log(5)*5**x

pol_10 = lambda x: 8*x**10 - 12*x**9 + 13*x**6 - 22*x**4 + 37*x**3 + 6*x
df_dx_pol10 = lambda x: 80*x**9 - 108*x**8 + 78*x**5 - 88*x**3 + 111*x**2 + 6
```

```python
# нахождение корней через sympy
x = sp.Symbol('x')
print(sp.solvers.solve((x+4)*(x+2)*(x-4), x))
print(sp.solvers.nsolve(sp.sin(2*x) + 5**x, -3), sp.solvers.nsolve(sp.sin(2*x) + 5**x, -2), 
      sp.solvers.nsolve(sp.sin(2*x) + 5**x, 0))
print(sp.solvers.nsolve(8*x**10 - 12*x**9 + 13*x**6 - 22*x**4 + 37*x**3 + 6*x, -2), 
     sp.solvers.nsolve(8*x**10 - 12*x**9 + 13*x**6 - 22*x**4 + 37*x**3 + 6*x, 1))
```

```python
x = np.linspace(-5, 5, 1000)
y1 = pol_3(x)
y2 = transcend(x)
y3 = pol_10(x)

fig, ax = plt.subplots(1, 3, figsize = (16, 4))
ax[0].plot(x, y1, color = 'green')
ax[0].plot([-5, 5], [0, 0], color = 'black', linestyle='--')
ax[0].set_title('Полином третьего порядка')
ax[0].set_xlabel("x")
ax[0].set_ylabel('f(x) = (x+4)(x+2)(x-4)')
ax[0].grid()

ax[1].plot(x[100:-400], y2[100:-400])
ax[1].plot([-4, 1], [0, 0], color = 'black', linestyle='--')
ax[1].set_title('Трансцендентная функция')
ax[1].set_ylabel('f(x) = sin(2x)+5^x')
ax[1].set_xlabel('x')
ax[1].grid()

ax[2].plot(x[380:-400], y3[380:-400], color = 'purple')
ax[2].plot([-1.5, 1], [0, 0], color = 'black', linestyle='--')
ax[2].set_title('')
ax[2].set_ylabel('f(x)')
ax[2].set_xlabel("x")
ax[2].grid()

```

```python
def bisection(func, range_, eps):
    start = timeit.default_timer()
    iters = 0
    
    while 1:      
        middle = (range_[0]+range_[1])/2
        
        if func(middle) * func(range_[0]) < 0:
            range_[1] = middle
        elif func(middle) * func(range_[1]) < 0:
            range_[0] = middle
        iters += 1 
        
        if abs(range_[0] - range_[1]) < 2*eps:
            middle = (range_[0]+range_[1])/2
            break
            
    t = start - timeit.default_timer()
    return middle, iters, t
```

```python
def secant_method(func, range_, eps):
    start = timeit.default_timer()
    iters = 0
    x_n = range_[1]
    x_n_0 = range_[0]
    
    while 1:
        x_n_1 = x_n - ((x_n - x_n_0) * func(x_n))/(func(x_n) - func(x_n_0))
        iters +=1
        if abs(x_n_1-x_n) < eps:
            break          
        else:
            x_n_0 = x_n
            x_n = x_n_1
    t = timeit.default_timer() - start
    
    return x_n_1, iters, t      
```

```python
def tangent_method(f, df_dx, range_, eps):
    start = timeit.default_timer()
    iters=0
    x_0 = (range_[0] + range_[1])/2
    x_1 = x_0 - f(x_0)/df_dx(x_0)
    while np.abs(x_0 - x_1) > eps:
        x_0 = x_1
        x_1 = x_0 - f(x_0)/df_dx(x_0)
        iters += 1
    t = timeit.default_timer() - start
    return x_1, iters, t
```
