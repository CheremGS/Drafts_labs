# Численное нахождение точек минимума градиентным методом

## 1 Условие

$$\normalsize f(x) = -ln(|x+4y| + 1) + 2y^2 + x^2$$

## 
## 2 Реализация градиентного метода

Сразу стоит отметить, что данная функция симметрична, относительно прямой $x+4y = 0$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# целевая функция f и ее производные 
f = lambda x, y: -np.log(np.abs(x + 4*y) + 1) + 2*y**2 + x**2
df_dx = lambda x, y: -1/(x + 4*y + 1) + 2*x if x + 4*y > 0 else 1/(-(x + 4*y) + 1) + 2*x
df_dy = lambda x, y: -4/(x + 4*y + 1) + 4*y if x + 4*y > 0 else 4/(-(x + 4*y) + 1) + 4*y
```

График функции от двух переменных:


Точки минумума(Wolfram Alpha):<br><br>
$$\normalsize min( f(x) ) \approx -0.672217\\ \normalsize(x_1, y_1) = (-0.186605, -0.373211);\\ \normalsize (x_2, y_2) = (0.186605, 0.373211).$$

Градиентный метод без оптимизации:

```python
def gradient_descent(func, deriv_x, deriv_y, x0, y0, eps = 0.001, alpha = 1):
    # обычный без оптимизации    
    x_k_1, y_k_1 = x0 - 2*eps, y0 - 2*eps
    mas_for_steps = []
    while 1:
        # частные производные в точках
        x_k_1 = x0 - alpha*(deriv_x(x0, y0))
        y_k_1 = y0 - alpha*(deriv_y(x0, y0))
         
        mas_for_steps.append([x0, y0])
        if np.abs(x_k_1 - x0) + np.abs(y_k_1 - y0) > eps/2:
            x0, y0 = x_k_1, y_k_1
        else:
            break
    return mas_for_steps
```

Градиентный метод с одномерной оптимизацией(метод дихотомии, другие методы не дают значимо отличающихся результатов):

```python
def gradient_descent(func, deriv_x, deriv_y, x0, y0, eps = 1e-3):
    
    mas_for_x = []
    mas_for_y = []
    
    while 1:
        # частные производные в точках
        x_0_der = deriv_x(x0, y0)
        y_0_der = deriv_y(x0, y0)
        
        # одномерная оптимизация    
        a = 0
        b = 1
        while 1:
            middle = (a + b)/2           
            if func(x0 - (middle+eps/2)*x_0_der, y0 - (middle+eps/2)*y_0_der) > func(x0 - (middle-eps/2)*x_0_der, y0 - (middle-eps/2)*y_0_der):
                b = middle
            else:
                a = middle
            
            if abs(a-b) < 2*eps:
                lam = middle
                break   
        print(lam)              
        # конец цикла оптимизации   
        
        x_k_1 = x0 - lam*x_0_der
        y_k_1 =  y0 - lam*y_0_der        
        mas_for_x.append(x0)
        mas_for_y.append(y0)
        if np.sqrt((x_k_1 - x0)**2 + (y_k_1 - y0)**2) > eps/2:
            x0, y0 = x_k_1, y_k_1
        else:
            break

    return mas_for_x, mas_for_y
```

## 
## 3 Результат выполнения программы и визуализация

 * $ \normalsize (x_0, y_0) = (3, 3); \; eps = 0.001$

```python
print('Коэффициент альфа(скорость спуска) на каждой итерации:')
x_value, y_value = gradient_descent(f, df_dx, df_dy, 3, 3, eps = 0.001)
```

```python
f_plot = lambda x, y: 2 * y ** 2 + x ** 2 - np.log(1 + np.abs(x + 4 * y))

color = 'OrRd_r'
x, y = np.mgrid[-1:1:50j, -1:1:50j]
z = f_plot(x, y)

fig = plt.figure(figsize = (10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap=color)
ax.view_init(30, 20)
ax.set_title('График поверхности вблизи точек минимума')
plt.show()


plt.figure(figsize = (10, 10))
x, y = np.mgrid[-5:5:50j, -5:5:50j]
z = f_plot(x, y)
C = plt.contour(x, y, z, 10, colors='khaki')

plt.contourf(x, y, z, 10, cmap=color)
plt.title('Итерации градиентного метода')
plt.clabel(C, fontsize=8)
plt.colorbar()

plt.plot(x_value, y_value, marker='o', markersize=5, markerfacecolor='black', lw=1, c='black')
plt.axis('square')
plt.show()
```
