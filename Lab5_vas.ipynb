## Условие

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
```

## Нахождение экстремали

```python
x = sp.symbols('x')
y = sp.Function('y')
euler_eqantion = sp.Eq(sp.diff(sp.diff(x*(sp.diff(y(x), x))**2 + y(x)*sp.diff(y(x), x), 
                                       sp.diff(y(x), x)), x) - sp.diff(x*(sp.diff(y(x), x))**2 + y(x)*sp.diff(y(x), x), y(x)))
```

```python
display('Уравнение Эйлера:', euler_eqantion)
```

```python
extr_field = sp.dsolve(euler_eqantion)
display('Полученое решение дифференциального уравнения:', extr_field)
```

## График поля экстремалей 

```python
x = np.linspace(0.2, 6, 200)
y = lambda a, b: a + b*np.log(x)
```

```python
y1 = y(0, 0.5)
y2 = y(0, 1)
y3 = y(0, 2)
y4 = y(0, 3)
y5 = y(0, 4)

y11 = y(0, -0.5)
y21 = y(0, -1)
y31 = y(0, -2)
y41 = y(0, -3)
y51 = y(0, -4)
```

```python
y6 = y(0.1, 1)
y7 = y(1, 1)
y8 = y(3, 1)
y9 = y(5, 1)
y10 = y(6, 1)

y61 = y(-3, 1)
y71 = y(-1, 1)
y81 = y(-2, 1)
y91 = y(-4, 1)
y101 = y(-6, 1)
```

```python
plt.figure(figsize = (12, 10))
plt.grid(True)
plt.plot([0, 0], [-8, 8], color = 'black')
plt.plot([-2, 7], [0, 0], color = 'black')

for y_i in [y1, y2, y3, y4, y5, y11, y21, y31, y41, y51]:
    plt.plot(x, y_i, color = 'green')
    
for y_i in [y6, y7, y8, y9, y10, y61, y71, y81, y91, y101]:
    plt.plot(x, y_i, color = 'blue')
    
plt.plot(x, y(0, 1), color = 'red', linewidth = 3)

display('Зеленые линии - центральное поле экстремалей')
display('Синие линии - собственное поле экстремалей')
```

Из граничных условий $\normalsize(y(1) = 0; y(e) = 1)$ экстремаль примет вид (красная линия на графике): $$\normalsize y(x) = ln(x)$$

Исследование на сильный (или слабый) экстремум:

<font size = 3> Первые два условия выполнены, осталось проверить 3 условие<br>
Функция Вейерштрасса имеет слудующий вид:

При данных условиях:<br>
$$\normalsize f(x, y, y') = x(y')^2 + yy'\\
\normalsize f(x, y, p) = xp^2 + yp\\
\normalsize f'_p(x, y, p) = 2xp + y$$

<font size = 3>После всех преобразований(приведение подобных и тп) над выражением мы получим следующее выражение:<br>
$$ \normalsize E(x, y, y', p) = x(y' - p)^2 \geq 0;$$ <br>Исходя из того условия, что наша экстремаль, соответствующая граничным условиям,<br> имеет вид $y = ln(x)$ (область определения $ x > 0$).<br>
Тогда мы можем заключить, что данная функция Вейерштрасса сохраняет знак в точках $(x, y)$, <br>близких к экстремали, а также при любых значениях $y'$(то есть во всех точках сильно окрестности), а значит мы имеем сильный минимум.

### Вывод: Была найдена экстремаль функционала, соответствующая граничным условиям  $y = ln(x)$. Данная экстремаль оказалась сильным минимумом, после исследования соответствующей функции Вейерштрасса.
