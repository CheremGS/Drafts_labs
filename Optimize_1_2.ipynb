### Метод деления отрезка пополам

<b>Описание метода:<br></b>
Метод  дихотомии (другое  название  -  метод  половинного  деления) предназначен  для  поиска  локального  экстремума  функции  на  заданном  интервале.   Исходный  отрезок   $[a, b]$  делят  пополам  точкой <b>$C$</b> координатой $$ C = \frac{a + b}{2}\$$  После  этого  находят  значения  функции  в  двух  пробных  точках,  находящихся  справа  и  слева  от  точки  с  на  расстоянии  $\frac{eps}{2}\$.  После  этого  из  двух  образовавшихся  половинок  отрезка  одну  отбрасывают,  а  имеено  ту,   в  которой  содержится  та  пробная  точка,  значение  функции в  которой  меньше  (при  поиске  максимума)  или  больше  (при  поиске  минимума).  Для  этого  находят  значения  функции  в  точках и сравнивают  их  между  собой  и  определяют  “отбрасываемую”  половину  отрезка.   Отбрасывание  производят  посредством  переноса  в  среднюю  точку  отрезка  (точку $с$)  другого  конца  отбрасываемого  отрезка  (т.е.  точки  $a$  или  $b$).  Так  производят  до  тех  пор,  пока  длина  оставшегося  отрезка  станет  меньшей  заданной  точности.  В  таком  случае  за  найденное  приближенное  значение  точки  достижения  функцией   экстремума  принимают  средину  этого  (последнего)  отрезка.<br>

<b>Реализация метода:</b>

```python
def bisection(f, l, r, eps = 0.001):
    l1, r1 = l, r
    mas_min = [l1, r1]
    l2, r2 = l, r
    mas_max = [l2, r2]
    while abs(r1 - l1) > eps: 
        middle_point = (r1 + l1)/2

        if f(middle_point + eps/2) > f(middle_point - eps/2):
            l1 = middle_point
        else:
            r1 = middle_point
        mas_min.append([l1, r1])
    l1_out = middle_point  
    
    while abs(r2 - l2) > eps: 
        middle_point = (r2 + l2)/2

        if f(middle_point + eps/2) < f(middle_point - eps/2):
            l2 = middle_point
        else:
            r2 = middle_point
        mas_max.append([l2, r2])
    l2_out = middle_point        
    
    return l1_out, l2_out
```

### Метод золотого сечения   

Метод  золотого  сечения   предназначен  для  поиска  локального  экстремума  функции  на  заданном  интервале.  Способ  достижения  цели  -  в  некотором  смысле  аналогичен  ранее  описанному  методу  дихотомии.   Исходный  отрезок   $[a, b]$   делят  на  три  части  точками $$\normalsize x_1 = a + \frac{b - a}{\Phi} ,  x_2 = b - \frac{b - a}{\Phi}; (\Phi = \frac{1 + \sqrt{5}}{2}\ \approx 1.618)$$.<br>  Очевидно,  что  $\large x_1 < x_2$.  После  этого  находят  значения  функции  в  крайних  точках  интервала  и  в  точках  $x_1$  и $x_2$.  Из  четырех  полученных  значений  функции  выбирают  минимальное  (при  поиске  минимума)  или  максимальное  (при  поиске  максимума).  Дальше  производят  “отбрасывание”  той  части  интервала  (одной  из  имеющихся  трех  частей),  которая  находится  наиболее  далеко  от  точки  экстремума. Это  достигается  переносом  точки  $a$  в  точку  $x_1$  или  точки  $b$  в  точку  $x_2$.  В  результате  проделанных  операций  длина  интервала  $[a, b]$  уменьшиться.  Теперь  к  оставшится  трем  точкам  $(a,  b,  x_1|x_2)$   добавим  четвертую  (недостающий  x)  по  формуле $a + b - x$  и  дальше  все  повторяем.  Так  продолжаем  до  тех  пор,  пока  длина  интервала  $[a, b]$ не  станет  меньше  заданной  точности.   В  этом  случае  за  точку  экстремума  примем  значение  одного  из  $x$.

<b>Реализация метода:

```python
def golden_section_methods(f, l, r, eps = 0.001):
    l1, r1 = l, r
    l2, r2 = l, r
    fi = (np.sqrt(5) + 1)/2
    while abs(r1 - l1) > eps: 
        I = (r1 - l1)/fi
        
        if f(r1) < f(l1):
            l1 += I
        else:
            r1 -= I
    if f(r1) < f(l1):
        out1 = r1
    else:
        out1 = l1
    
    while abs(r2 - l2) > eps: 
        I = (r2 - l2)/fi
        
        if f(r2) > f(l2):
            l2 += I
        else:
            r2 -= I
    
    if f(r2) > f(l2):
        out2 = r2
    else:
        out2 = l2
    
    return out2, out1
```

### Метод парабол

Метод парабол предназначен для поиска локальных экстремумом. Суть метода заключается в построении параболл по трем точкам, две из которых являются концами отрезка. При этом для определения коэфициентов в уравнении $ ax^2 + bx + c = 0 $ требуется решить систему из трех уравнений(т.к. мы проводим параболлу через три точки, где известны координаты). После построения параболлы необходимо найти минимум(максимум) этой параболлы по формуле $\normalsize x = \frac{-b}{2a}$. Затем проводим следующую параболу, только теперь третья точка - точка минимума(максимума) предыдущей параболы. Так выполняем построение парабол, пока разница между двумя точками на разных итерациях больше заданной точности.

<b>Реализация метода:

```python
def parabolic_method(f, l, r, eps = 0.001):
    x2 = (r + l)/3
    while True:
        mas = np.linalg.solve(np.array([np.array([l**2, l, 1]), 
                             np.array([x2**2, x2, 1]), 
                             np.array([r**2, r, 1])]), np.array([f(l), f(x2), f(r)]))
        u = - mas[1]/2/mas[0]
        
        if abs(u - x2) < eps:
            break       
        if r - u < 0: 
            x2 = r
            break
        if u - l < 0:
            x2 = l
            break       
        x2 = u
        
    x1 = (r + l)/3
    while True:
        mas = np.linalg.solve(np.array([np.array([l**2, l, 1]), 
                             np.array([x1**2, x1, 1]), 
                             np.array([r**2, r, 1])]), np.array([f(l), f(x1), f(r)]))
        u = mas[1]/2/mas[0]
        
        if abs(u - x1) < eps:
            break       
        if r - u < 0: 
            x1 = r
            break
        if u - l < 0:
            x1 = l
            break       
        x1 = u

    return x1, x2
```

```python
out = pd.Series()
s = timeit.default_timer()
r1 = bisection(f, left_bound, right_bound)
e = timeit.default_timer()
e1 = e - s
out['Дихотомия'] = r1

s = timeit.default_timer()
r2 = golden_section_methods(f, left_bound, right_bound)
e = timeit.default_timer()
e2 = e - s
out['Золотое сечение'] = r2

s = timeit.default_timer()
r3 = parabolic_method(f, left_bound, right_bound)
e = timeit.default_timer()
e3 = e - s
out['Метод параболл'] = r3
```

### 2 <u>Метод секущих</u>

В этом методе задача поставлена, также как и в предыдущем методе - поиск корней уравнения $F'(x) = f(x) = 0$. Однако, теперь приближение будет производится на интервале $[a, b]$ при помощи отделения отрезка по формуле $$x_{k} = a_k - F'(a_k)\frac{a_k - b_{k}}{F'(a_k) - F'(b_{k})}$$
Эта точка является точкой пересечения графика $F'(x)$ и осью $Ox$. Но это при случае, если график пересекает ось $Ox$, другими словами, если $ F'(a_k)F'(b_k) < 0$. В противном же случае(функция монотонна, так как производная не меняет знак) есть два варианта:
* $x_k = a_k$, при условии $ F'(a_k) > 0; F'(b_k) > 0$
* $x_k = b_k$, при условии $ F'(a_k) < 0; F'(b_k) < 0$

В качестве начального интервала не следует брать точки, находящиеся слишком рядом с нулем, так как алгоритм будет сходится к очень медленной скоростью.

Графическая интерпретация метода


### Реализация метода:

```python
def secant(f, eps = 1e-3):
    mas2 = [] 
    
    a = 1.1
    b = 6
    x = a
    x_p = 100
    i = 1
    while abs(x_p - x) >= eps or f(a)*f(b) > 0:
        x_p = x
        x = a - f(a) * (a - b)/(f(a) - f(b))
        if f(a)*f(x) > 0:
            a = x
        else:
            b = x
        mas2.append(x)
        i += 1
    mas2.append(x)
    
    return mas2
```
